```python
import pandas as pd
import requests

url = 'https://data.coa.gov.tw/Service/OpenData/FromM/AquaticTransData.aspx'
response = requests.get(url)

if response.status_code == 200:
    data = response.json()

df = pd.DataFrame(data)
df.to_csv('AquaticTransData.csv', index = False, encoding='utf-8')
df.to_excel('AquaticTransData.xlsx')
df1 = pd.read_csv('AquaticTransData.csv')

df1.head(10)
```
